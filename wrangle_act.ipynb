## Clean
In this section, I will start to execute cleaning for the issues found. Please note that these are not all the issues.

````python
tweet_clean = tweet.copy()
image_predictions_clean = image_predictions.copy()
json_tweets_clean = json_tweets.copy()
````

**`retweeted_status_id`, `retweeted_status_user_id`, `retweeted_status_timestamp` show 181 records, this means that these records are retweets from our original tweets gathered (i.e. a duplicate of the original tweet).**

[A good hack here](https://stackoverflow.com/questions/26535563/querying-for-nan-and-other-names-in-pandas) to use, I will use it to remove the retweeted records and then drop the `retweeted_status_id`, `retweeted_status_user_id`, `retweeted_status_timestamp` columns.

````python
#I will remove all 181 records from this dataset
tweet_clean = tweet_clean.query('retweeted_status_id != retweeted_status_id or retweeted_status_user_id != retweeted_status_user_id or retweeted_status_timestamp != retweeted_status_timestamp')
#next I will drop these columns
tweet_clean = tweet_clean.drop(['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis = 1)
````

**`in_reply_to_status_id`, `in_reply_to_user_id` all have a majority of null data, also the source column not needed.**

Remove these columns from the dataset.

````python
#just drop these columns, not needed anyway
tweet_clean = tweet_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'source'], axis = 1)
````

**`timestamp` format is an object not a datetime.**

Convert type from object to datetime.

````python
tweet_clean.timestamp = pd.to_datetime(tweet_clean.timestamp)
````

**`rating_numerator`, `rating_denominator` both have unlogical values. The numerator has very high values up to 1776, while there are rows that have the denominator more than 10 (which is not correct according to the info in hand).**

Just filter our any denominator value that is either less than or more than 10.

_Note that for the numerator I will leave as it is, the rating here did not specify a threshold to it, and this is the key value for the twitter account. Any extreme numerator rating is just a reflection of the social media hype :)._

````python
#The denominator
tweet_clean = tweet_clean[tweet_clean['rating_denominator'] == 10]
````

**`rating_numerator`, `rating_denominator` should be combined together as the twitter rating, this also means that these data are categorial.**

````python
#Do the twitter_rating column
tweet_clean['twitter_rating'] = tweet_clean.rating_numerator.map(str) + '/' + tweet_clean.rating_denominator.map(str)
tweet_clean.twitter_rating = tweet_clean.twitter_rating.astype('category')
````

````python
tweet_clean = tweet_clean.drop(['rating_numerator','rating_denominator'], axis = 1)
````

**Records have invalid names, they are easily detected as they start with a lower case letter. These namings should be replaced with a more appropriate description as 'None'.**

The trend of the invalid names is starting with a lowercase letter can be replaced as 'None', with the help of [this link](https://stackoverflow.com/questions/50599934/removing-rows-from-dataframe-whose-first-letter-is-in-lowercase) and the `.loc` function.

````python
#Change the invalid names to 'None'
tweet_clean.loc[tweet_clean['name'].str.contains('^[a-z]+'),'name'] = 'None'
````

**Inconsistency in the 3 breed predictors in terms of casing.**

Make all of the data values as lowercase.

````python
#convert all values in p1, p2 and p3 to lower cases for ease of use
image_predictions_clean.p1 = image_predictions_clean.p1.str.lower()
image_predictions_clean.p2 = image_predictions_clean.p2.str.lower()
image_predictions_clean.p3 = image_predictions_clean.p3.str.lower()
````

**Some breed namings are invalid.**

Look for the first predicitons which resulted **True** in `p1`, `p2`, and `p3`, append them to a new column, and the same for the corresponding confidence level `p1_conf`, `p2_conf`, `p3_conf`. There are multiple ways of doing this, I choose to cluster each column alone and use the [np.select](https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column) to create each new column `prediction` and `confidence_lvl` respectively.

````python
#First the prediction
conditions = [image_predictions_clean['p1_dog'] == True,image_predictions_clean['p2_dog'] == True,image_predictions_clean['p3_dog'] == True]
returns = [image_predictions_clean['p1'],image_predictions_clean['p2'],image_predictions_clean['p3']]
image_predictions_clean['prediction'] = np.select(conditions,returns, default = 'None')
````

````python
#Second the confidence level
conditions = [image_predictions_clean['p1_dog'] == True,image_predictions_clean['p2_dog'] == True,image_predictions_clean['p3_dog'] == True]
returns = [image_predictions_clean['p1_conf'],image_predictions_clean['p2_conf'],image_predictions_clean['p3_conf']]
image_predictions_clean['confidence_lvl'] = np.select(conditions,returns, default = 0)
````

````python
#Now to drop the unnecessary columns
image_predictions_clean = image_predictions_clean.drop(['p1','p1_conf','p1_dog','p2','p2_conf','p2_dog','p3','p3_conf','p3_dog',], axis = 1)
````

````python
#Change the prediction to a categorial type instead of a string, breeds are categories
image_predictions_clean.prediction = image_predictions_clean.prediction.astype('category')
````

**Variables doggo, floofer, pupper, puppo should be in one column.**

Note the presence of 2 stages containing values for one row inside the dataset. I was influenced by the work of [Abdul](https://github.com/Abdul-al5aldi/DAND_Udacity_5/blob/master/wrangle_act.ipynb) and the use of the [.loc](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) function.

````python
#Change None to just an empty string, I'll need to do so to have a lean clean set of data
tweet_clean.loc[tweet_clean.doggo == 'None','doggo'] = ''
tweet_clean.loc[tweet_clean.pupper == 'None','pupper'] = ''
tweet_clean.loc[tweet_clean.puppo == 'None','puppo'] = ''
tweet_clean.loc[tweet_clean.floofer == 'None','floofer'] = ''
#Then combine them to one column, and concat the dual values properly
tweet_clean['dog_stage'] = tweet_clean.doggo + tweet_clean.floofer + tweet_clean.pupper + tweet_clean.puppo
tweet_clean.loc[tweet_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo,pupper'
tweet_clean.loc[tweet_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo,puppo'
tweet_clean.loc[tweet_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo,floofer'


tweet_clean.loc[tweet_clean.dog_stage == '', 'dog_stage'] = 'None'
````

````python
#Now drop the 4 columns
tweet_clean = tweet_clean.drop(['doggo','puppo','pupper','floofer'], axis = 1)
````

**The 4 variables should be formatted as a category (This is related to the first tidiness issue also).**

I will convert these variables to category after combining them to a single column.

````python
tweet_clean.dog_stage = tweet_clean.dog_stage.astype('category')
````

**The json_tweets dataset can be merged to the tweet dataset. Same goes for the image_predictions dataset.**

The final step in the cleaning process is to merge the datasets into one, looking at the 3, all data inside are qualified to be consolidated to one for the analysis phase.

I will store the cleaned data inside an SQLite database, and the combined data in `twitter_archive_master.csv`. A [helpful link](https://stackoverflow.com/questions/50803109/how-to-store-pandas-dataframe-in-sqlite-db)

````python
# First storing in an SQLite Db
conn = sqlite3.connect('C:\\Users\\ASUS\\OneDrive\\Courses\\Business Analytics\\Data Analyst Nanodegree\\Part04DataWrangling\\Project05WrangleAndAnalyzeData\\rate_your_dog.sqlite')
````

````python
# Load your clean datasets
tweet_clean.to_sql('tweet_clean', conn, if_exists='replace', index=False)
image_predictions_clean.to_sql('image_predictions_clean', conn, if_exists='replace', index=False)
json_tweets_clean.to_sql('json_tweets_clean', conn, if_exists='replace', index=False)
````

````python
# Merging the json_tweets_clean and tweet_clean
twitter_archive_master = tweet_clean.merge(json_tweets_clean, how = 'left', left_on = 'tweet_id', right_on = 'id')
````

````python
# Merging the image_predictions_clean
twitter_archive_master = twitter_archive_master.merge(image_predictions_clean, how = 'left', left_on = 'tweet_id', right_on = 'tweet_id')
````

````python
# Final drop of id columns
twitter_archive_master = twitter_archive_master.drop(['id'], axis = 1)
````

````python
# A final clean to our dataset
twitter_archive_master.confidence_lvl = twitter_archive_master.confidence_lvl.fillna(0)
twitter_archive_master.prediction = twitter_archive_master.prediction.fillna('None')
````

````python
# Now load to a .csv file
twitter_archive_master.to_csv('C:\\Users\\ASUS\\OneDrive\\Courses\\Business Analytics\\Data Analyst Nanodegree\\Part04DataWrangling\\Project05WrangleAndAnalyzeData\\twitter_archive_master.csv', index=False)
````
